```python

for i in range(5):

    last_chain_seen = "A"
    
    offset = 1000
    out = []
    with open(f"../final/HO/HO_{i}.pdb") as f:
        for line in f.readlines():
            
            if line[:4] == "ATOM" and line[21] != "A":
                resid = int(line[22:26]) + offset
                out_line = f"{line[:21]}A{resid:4d}{line[26:]}"
                out.append(out_line)
            else:
                if line.startswith("TER") or line.startswith("END"):
                    continue

                resid = int(line[22:26])
                out_line = f"{line[:21]}A{resid:4d}{line[26:]}"
                
                out.append(out_line)
    
        out.append("TER\n")
        out.append("END\n")
                
    with open(f"HO_{i}.pdb","w") as f:
        f.writelines(out)
        
```

```python
import pandas as pd

for i in range(5):

    df = pd.read_csv(f"../final/L/L_{i}.csv")
    df = df.drop(columns="Unnamed: 0")
    df["Mutation"] = [f"{m[0]}{int(m[1:-1])+1}{m[-1]}" for m in df["Mutation"]]
    df["pos"] = df["pos"] + 1
    
    df.to_csv(f"../final/L/L_{i}.csv",index=False)
        
```
