```python
import tfscreen
from tfscreen.plot import heatmap
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from tfscreen.util import check_columns
```

## Make heat maps

Make various heat maps.

```python
theta_file = "/Users/harmsm/Desktop/keep/theta_df.csv"

# Load dataframe and make genotype into categorical
df = tfscreen.util.read_dataframe(theta_file)
df = tfscreen.genetics.set_categorical_genotype(df)
df = tfscreen.genetics.expand_genotype_columns(df)
```

```python
single_df = tfscreen.genetics.get_single_with_wt(df,["titrant_conc"])
```

```python

from tfscreen.plot.heatmap import aa_v_res_heatmap

to_plot_df = single_df[single_df["titrant_conc"] == 0.]

fig, ax = aa_v_res_heatmap(to_plot_df,
                           value_column="theta_est",
                           heatmap_kwargs={"vlim":[0.7,1.9]})
```

```python
from tfscreen.plot.heatmap import aa_v_titrant_heatmap

sub_df = single_df[single_df["resid"] == 30]
heatmap_kwargs = {"x_axis_kwargs":{"max_num_ticks":3}}

fig, ax = aa_v_titrant_heatmap(sub_df,
                               titrant_column="titrant_conc",
                               value_column="theta_est",
                               heatmap_kwargs=heatmap_kwargs)
```

```python
from tfscreen.plot.heatmap import epistasis_heatmap

sub_df = df[df["titrant_conc"] == 0 ]
fig, ax = epistasis_heatmap(sub_df,29,98,
                            value_column="theta_est",
                            heatmap_kwargs={"vlim":[-4,4]})
```
